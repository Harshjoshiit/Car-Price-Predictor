## Model

```python
X=car[['name','company','year','kms_driven','fuel_type']]
y=car['Price']
```

```python
X=X.reset_index(drop='true')
y=y.reset_index(drop='true')
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
ohe=OneHotEncoder() 
ohe.fit(X[['name','company','fuel_type']]) 

# What OneHotEncoder does:It converts each category into a binary column (0 or 1).
# 🤔 Why do we need OneHotEncoder (OHE)?
# Machine learning models (like linear regression, decision trees, etc.) cannot understand text directly. They work with numbers only.
```

## Creating column transformer to transform categorical columns

❓1. Do we only need to transform categorical columns, not integer columns?
✅ Answer:
Yes — usually, only categorical columns need transformation like OneHotEncoding, because:

ML models can't understand text like "Maruti" or "Petrol", so we convert them to numbers.

Integer columns (like year, kms_driven, price) are already numeric — so models can use them directly.

```python
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),
                                    remainder='passthrough')
```

```python
lr = LinearRegression()
```

```python
pipe=make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
pipe.predict(pd.DataFrame(data=[['Hyundai i20 Magna','Hyundai',2013,35000,'Petrol']],columns=X_test.columns))
```

```python
scores=[]
for i in range(1000):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
```

```python
np.argmax(scores) 
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
r2_score(y_test,y_pred)
```

```python
pipe.predict(pd.DataFrame(data=[['Tata Indigo CS','Tata',2017,50000,'Diesel']],columns=X_test.columns))
```
